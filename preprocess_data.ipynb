```python
import pandas as pd
import numpy as np
import json
```

# QUICK VIEW OF BOOK DATAFRAME


```python
book_df = pd.read_csv('book.csv')
book_df
```

# CHECKING TYPE OF SPECIFICATIONS FEATURE

```python
print(book_df['specifications'][0])
print(type(book_df['specifications'][0]))
```

# CONVERT SPECIFICATIONS INTO LIST DATATYPE TO MAKE IT EASIER TO PREPROCESS


```python
book_df['specifications'] = book_df['specifications'].apply(lambda x: eval(x))

```

```python
print(type(book_df['specifications'][0]))
```

# CHECKING DUPLICATE IN BOOK_ID COLUMNS

```python
duplicate_row = book_df[book_df['id'].duplicated()]
duplicate_row
```

```python
duplicate_name = book_df[book_df['name'].duplicated()]
duplicate_name
```

```python
book_df = book_df.drop_duplicates(subset=['name'])
book_df = book_df.drop_duplicates(subset=['id'])
```

# FLATTEN 'SPECIFICATIONS' FEATURE



```python
book_df['attributes'] = book_df['specifications'].apply(lambda x: x[0]["attributes"])
book_df
```

```python
#check the number of attributes in each book
book_df['n_attr'] = book_df['attributes'].apply(lambda x : len(x))
print(book_df['n_attr'].unique())

```

```python
book_df
```

# add pulisher_vn attribute to dataframe

```python
book_df['publisher_vn'] = book_df['attributes'].apply(lambda x : x[0]['value'])

def find_book_cover(x):
    for item in x:
        if item['name'] == 'book_cover' :
            return item['value']
    return "N\A"
def find_number_of_page(x):
    for item in x:
        if item['name'] == 'number_of_page' :
            return item['value']
    return "N\A"
def find_manufacturer(x):
    for item in x:
        if item['name'] == 'manufacturer' :
            return item['value']
    return "N\A"
book_df['book_cover'] = book_df['attributes'].apply(find_book_cover)
book_df['number_of_page'] = book_df['attributes'].apply(find_number_of_page)
book_df['manufacturer'] = book_df['attributes'].apply(find_manufacturer)

```

```python
book_df
```

```python
cleaned_book_df = book_df.drop(columns=["specifications","attributes","n_attr"])
cleaned_book_df.to_csv('cleaned_book.csv', index=False)
```
