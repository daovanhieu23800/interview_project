```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
book_df = pd.read_csv('cleaned_book.csv')
book_df
```

# SEE HOWMANY PUSHLISHER IN VN, AND THE AFFECT OF ITS TO PRICE

```python
print(len(book_df['publisher_vn'].unique()))
print(book_df['publisher_vn'].unique())
```

```python
label_counts = book_df['publisher_vn'].value_counts()
fig, ax = plt.subplots(figsize=(10, 6))  # Adjust the figure size as per your preference
label_counts.plot.bar(ax=ax)
ax.set_xticklabels(label_counts.index, rotation=45, ha='right')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Bar Chart - Count of Categories')

# Show the plot
plt.tight_layout()  # To ensure the labels and title are properly displayed
plt.show()
```
